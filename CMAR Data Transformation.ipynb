```python
# Import libraries
import pandas as pd
from sodapy import Socrata
```

```python
client = Socrata("data.novascotia.ca", 'cx4XloH6tDgX8ZhFFmtAuxtMc')
```

```python
# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("eb3n-uxcb", limit=6000000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

```python
results_df
```

```python
subset_df = results_df.iloc[0:2000]
subset_df.loc[:, ('timestamp')] = pd.to_datetime(subset_df['timestamp'])
subset_df.loc[1] = subset_df.loc[0]
subset_df.loc[1:1, ('variable')] = 'salinity'
# typed_subset_df = subset_df.infer_objects()
subset_df
```

```python
grouped_df = subset_df.groupby(['waterbody', 'station', 'lease', 'latitude', 'longitude', 'deployment_period', 'timestamp', 'sensor', 'depth'])
```

```python
def merge_timestamps(group):
    non_unique = ['variable', 'value']
    
    new_df = pd.DataFrame()
    
    # Copy over all of the unique columns     
#     for col in group:
#         if col not in non_unique:
#             new_df[col] = group[col].unique()
    
    # Handle the non-unique columns
    # Create a column for each variable, store the associated value
    # If there are multiple, different values for the same variable, at the same time, raise an error     
    sci_variable_list = group['variable'].unique()
    for variable in sci_variable_list:
        if len(group['value'].unique()) > 1:
            raise("Multiple values for the same variable at timestamp: %s" % new_df['timestamp'])
        else:
            new_df[variable] = group['value'].unique()
    return new_df
```

```python
merged_df = grouped_df.apply(merge_timestamps)
```
